Lets create a heatmap that would showcase how the given label and how much percentage of data is having 1 

```python
import seaborn as sns;
import matplotlib.pyplot as plt;
```

```python
percentagePixel = [];
for labelIndex in digit_df["label"].unique().tolist():
    percentagePixel.append([labelIndex] + (digit_df.loc[digit_df["label"]==labelIndex, pixelCols].sum(axis=0)/digit_df.shape[0]).values.tolist())
    
```

```python
percentagePixel = pd.DataFrame(percentagePixel, columns=["label"] +pixelCols )
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(percentagePixel)
```

Lets try and see what would be the overall shapes look like

```python
sorted(digit_df["label"].unique().tolist())
```

```python
fig, axs = plt.subplots(2, 5)
for labelIndex in sorted(digit_df["label"].unique().tolist()):
    #print(labelIndex)
    if(labelIndex < 5):
        sns.heatmap(percentagePixel.loc[percentagePixel["label"]==labelIndex, pixelCols].values[0].reshape(28, 28), ax=axs[0][labelIndex])
    else:
        sns.heatmap(percentagePixel.loc[percentagePixel["label"]==labelIndex, pixelCols].values[0].reshape(28, 28), ax=axs[1][labelIndex-5])
plt.show()
    
```

## KMeans

```python
# To perform KMeans clustering 
from sklearn.cluster import KMeans
```

```python
# Kmeans with K=5
model_clus5 = KMeans(n_clusters = 10, max_iter=50)
model_clus5.fit(digit_df[pixelCols])
```

```python
model_clus5.labels_
```

```python
digit_df["predicted"] = model_clus5.labels_
```

```python
digit_df[["label", "predicted"]].head(20)
```

## train, test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digit_df[pixelCols], digit_df["label"], test_size=0.33, random_state=42)
```

```python
X_train.shape
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(multi_class='multinomial', solver='lbfgs')
```

```python
lr.fit(X_train, y_train)
```

```python
y_predict = lr.predict(X_test)
```

```python
y_predict
```

```python
y_test
```

```python
accuracy_score(y_test, y_predict)
```

```python
confusion_matrix(y_test, y_predict)
```

## Multinomial Naive Bayes

```python
# building a multinomial NB model
from sklearn.naive_bayes import MultinomialNB
```

```python
# instantiate NB class
mnb=MultinomialNB()
```

```python
# fitting the model on training data
mnb.fit(X_train, y_train)
```

```python
prob_predict = mnb.predict(X_test)
```

```python
proba = mnb.predict_proba(X_test)
```
